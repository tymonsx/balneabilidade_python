```python
from statsmodels.tsa.statespace.sarimax import SARIMAXResults
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loaded = SARIMAXResults.load('model.pkl')
loaded
```

```python
import datetime as dt
pred_uc = loaded.get_forecast(steps=84)
index_date = pd.date_range(dt.datetime(2020,7,26) , periods = 84, freq = 'W')
forecast_series = pd.Series(list(pred_uc.predicted_mean), index = index_date)
pred_ci = pred_uc.conf_int()
ax =forecast_series.plot(label='predito',figsize=(14, 4))
ax.fill_between(forecast_series.index,
 pred_ci.iloc[:, 0],
 pred_ci.iloc[:, 1], color='k', alpha=.1)
ax.set_xlabel('Data')
ax.set_ylabel('Enterococos')
plt.legend()
plt.show()
```
